```python
model = Top2Vec.load("t2v_h")
```

```python
topic_num=model.get_num_topics()

topic_num
```

```python
topic_words, word_scores, topic_nums = model.get_topics(1)

topic_words
```

```python
'''==================PARAMETER TO MODIFY=================='''

num_topics_to_get=15 # the num of topics you want to get

'''======================================================='''

levels=model.hierarchical_topic_reduction(topic_num//num_topics_to_get)
```

```python
low = len(levels)//4
mid = len(levels)//2
high = len(levels)*3//4

low_res, mid_res, high_res, min_res, max_res = [], [], [], [], []

topic_words, word_scores, topic_nums = model.get_topics()

for v in levels[mid]:
    mid_res.append(topic_words[v])

for v in levels[low]:
    low_res.append(topic_words[v])

for v in levels[high]:
    high_res.append(topic_words[v])

for v in levels[0]:
    min_res.append(topic_words[v])

for v in levels[-1]:
    max_res.append(topic_words[v])
```
